本練習的目的是探索差異和差異對信號的影響。創建一個三角波並將其繪製。應用diff並繪製結果。計算三角波的頻譜，進行微分並繪製結果圖。將頻譜轉換回波形並繪製。此波的差異效果和差異效果之間是否存在差異？

```python
import numpy as np
import matplotlib.pyplot as plt

from thinkdsp import decorate
```

```python
#建三角波
from thinkdsp import TriangleSignal 

in_wave = TriangleSignal(freq=50).make_wave(duration=0.1, framerate=44100)
in_wave.plot()
decorate(xlabel='Time (s)')
```

三角波的diff是方波，這說明了為什麼方波中的諧波像1 /𝑓2那樣下降，卻像1 /𝑓那樣下降。

```python
#diff並繪製結果
out_wave = in_wave.diff()
out_wave.plot()
decorate(xlabel='Time (s)')
```

當我們採用頻譜導數時，我們會在不連續點周圍出現“振鈴”
從數學上講，問題在於三角波的導數在三角點處是不確定的。

```python
# 頻譜轉換回波形
out_wave2 = in_wave.make_spectrum().differentiate().make_wave()
out_wave2.plot()
decorate(xlabel='Time (s)')
```
