本練習的目的是探索信號累積和整合的影響。創建一個方波並將其繪製。應用求和並繪製結果。計算方波的頻譜，進行積分並繪製結果圖。將頻譜轉換回波形並繪製。此波的累積效應和積分效應之間是否有差異？

```python
import numpy as np
import matplotlib.pyplot as plt

from thinkdsp import decorate
```

```python
# 建方波
from thinkdsp import SquareSignal

in_wave = SquareSignal(freq=50).make_wave(duration=0.1, framerate=44100)
in_wave.plot()
decorate(xlabel='Time (s)')
```

方波的累積和是三角波。

```python
out_wave = in_wave.cumsum()
out_wave.plot()
decorate(xlabel='Time (s)')
```

頻譜積分也是一個三角波，儘管幅度非常不同。

```python
spectrum = in_wave.make_spectrum().integrate()
spectrum.hs[0] = 0
out_wave2 = spectrum.make_wave()
out_wave2.plot()
decorate(xlabel='Time (s)')
```

如果我們對兩個波進行無偏並歸一化，則它們在視覺上是相似的。

```python
out_wave.unbias()
out_wave.normalize()
out_wave2.normalize()
out_wave.plot()
out_wave2.plot()
```

它們在數值上相似，但精度只有3位左右。

```python
out_wave.max_diff(out_wave2)
```
