```python
from thinkdsp import Sinusoid
from thinkdsp import normalize, unbias
import numpy as np

class HYJ_Sawtooth(Sinusoid):
    
    def evaluate(self, ts):
        cycles = self.freq * ts + self.offset / np.pi / 2
        frac, _ = np.modf(cycles)
        ys = normalize(unbias(frac), self.amp)
        return ys
```

.2

```python
signal=HYJ_Sawtooth(freq=1000,amp=10)
signal.plot()
```

.3

```python
sawtooth = HYJ_Sawtooth().make_wave(duration=0.5, framerate=40000)
sawtooth.make_audio()
```
